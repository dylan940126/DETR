```python
import torch
from torch.utils.data import DataLoader
from torchvision.datasets.coco import CocoDetection
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    # transforms.Resize((256, 256)),
    transforms.ToTensor()
])

coco_train = CocoDetection(root='coco/images', annFile='coco/annotations/instances_val2017.json', transform=transform)
train_loader = DataLoader(coco_train, batch_size=1, shuffle=True)
```

```python
for i, (img, target) in enumerate(train_loader):
    print(i, type(img), type(target))
    if i == 1:
        break
```

```python
for img, target in train_loader:
    plt.imshow(img[0].permute(1, 2, 0))
    coco_train.coco.showAnns(target)
    plt.show()
```
